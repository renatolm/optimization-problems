Os centros 1 e 4 operam 16 horas por dia. Os centros 2 e 3 operam 12 horas por dia. A refinaria possui a capacidade de transportar somente 2.500 L/dia, pois seu oleoduto está em manutenção. Formule matematicamente o problema de otimizar a produção dos dois tipos de óleo.

## Modelagem matemática

### Conjuntos

Tipos de óleo: $O = \{I, II\}$

Centros de processamento: $C = \{1, 2, 3, 4, 21 \}$

_Obs_: Para facilitar a modelagem, vamos representar a segunda passagem pelo Centro 2 no Fluxo Tipo I como se fosse um novo centro de processamento denotado por $21$.

### Parâmetros

- $cap_{i,j}$: Capacidade de processamento do óleo $i \in O$ no centro $j \in C$ em L/h
- $rec_{i,j}$: Recuperação do óleo $i \in O$ no centro $j \in C$ em %
- $cproc_{i,j}$: Custo de processamento do óleo $i \in O$ no centro $j \in C$ em R$/h
- $cmat_{i}$: Custo da matéria-prima para produzir o óleo $i \in O$ em R$/L
- $preco_{i}$: Preço do venda do óleo $i \in O$ em R$/L
- $vdiaria_{i}$: Venda diária máxima do óleo $i \in O$ em L
- $horas_{j}$: Tempo de funcionamento diário do centro $j \in C$ em horas

### Variáveis de decisão

- $x_{i,j}$: Quantidade de óleo $i \in O$ processada no centro $j \in C$ em L

### Função objetivo

Para definir a função objetivo desse problema, precisamos primeiramente definir qual é a quantidade total de cada tipo de óleo produzida em cada fluxo, já que há uma perda (representada pela % de recuperação) a cada etapa de processamento.

- Fluxo Tipo I: $rec_{I,21} \cdot x_{I,21} = 0.8 \cdot x_{I,21}$
- Fluxo Tipo I - Alternativo: $rec_{I,3} \cdot x_{I,3} = 0.75 \cdot x_{I,3}$
- Fluxo Tipo II: $rec_{II,4} \cdot x_{II,4} = 0.8 \cdot x_{II,4}$

Também é interessante já definir os custos de produção, que são divididos entre o custo de matéria-prima de cada tipo de óleo e o custo de processamento em cada centro.

- Custo total de matéria prima: $cmat_I \cdot x_{I,1} + cmat_{II} \cdot x{II,1}$
- Custo de processamento em cada centro: $\sum\limits_{i \in O} \sum\limits_{j \in C} cproc_{i,j} \cdot x_{i,j}$

_Obs_: Não é possível processar qualquer tipo de óleo em qualquer centro de processamento, mas ajustaremos isso nas restrições do problema.

Agora podemos escrever a função objetivo.

- Maximizar o lucro total da refinaria (vendas - custos)
$$ \max Z = \underbrace{\left( preco_{I} \cdot (0.8 \cdot x_{I,21} + 0.75 \cdot x_{I,3} ) + preco_{II} \cdot 0.8 \cdot x_{II,4} \right)}_{\text{receita das vendas}} - \underbrace{\left( cmat_I \cdot x_{I,1} + cmat_{II} \cdot x_{II,1} \right)}_{\text{custo matéria-prima}} - \underbrace{\left( \sum\limits_{i \in O} \sum\limits_{j \in C} \frac{cproc_{i,j} \cdot x_{i,j}}{cap_{i,j}} \right)}_{\text{custo processamento}}

### Restrições

- Capacidade de processamento de cada centro, para cada tipo de óleo:
$$ x_{i,j} \leq cap_{i,j} \cdot horas_{j}, \quad \forall i \in O, \forall j \in C $$

- Fluxos e % de recuperação:
$$ x_{I,2} = 0.9 \cdot x_{I,1} $$
$$ x_{I,4} = 0.95 \cdot x_{I,2} $$
$$ x_{I,21} + x_{I,3} = 0.85 \cdot x_{I,4} $$
$$ x_{II,3} = 0.9 \cdot x_{II,1} $$
$$ x_{II,4} = 0.85 \cdot x_{II,3} $$
$$ x_{II,2} = 0 $$
$$ x_{II,21} = 0 $$

- Venda diária máxima:
$$ 0.8 \cdot x_{I,21} + 0.75 \cdot x_{I,3} \leq 1700 $$
$$ 0.8 \cdot x_{II,4} \leq 1500 $$

- Capacidade de transporte (do produto final):
$$ 0.8 \cdot x_{I,21} + 0.75 \cdot x_{I,3} + 0.8 \cdot x_{II,4} \leq 2500 $$

- Restrições de não-negatividade:
$$ x_{i,j} \geq 0, \quad \forall i \in F, \forall j \in C $$

## Modelagem e solução computacional

Agora vamos implementar a modelagem feita utilizando a biblioteca pyomo, e calcular a solução do problema.

```python
import pyomo.environ as pyo

model = pyo.ConcreteModel()
```

```python
# 1) Conjuntos
model.Oleos = pyo.Set(initialize=["I", "II"])
model.Centros = pyo.Set(initialize=[1,2,3,4,21])
```

```python
# 2) Parâmetros
capacidade = {('I', 1): 300,
              ('I', 2): 450,
              ('I', 4): 250,
              ('I', 21): 400,
              ('I', 3): 350,
              ('II', 1): 500,
              ('II', 3): 480,
            ('II', 4): 400} # litros por hora
model.cap = pyo.Param(model.Oleos,model.Centros, initialize=capacidade, default=1)

recuperacao = {('I', 1): 0.9,
              ('I', 2): 0.95,
              ('I', 4): 0.85,
              ('I', 21): 0.8,
              ('I', 3): 0.75,
              ('II', 1): 0.9,
              ('II', 3): 0.85,
            ('II', 4): 0.8}   # a cada 1 litro  
model.rec = pyo.Param(model.Oleos,model.Centros, initialize=recuperacao, default=0)

custo_processamento = {('I', 1): 150,
              ('I', 2): 200,
              ('I', 4): 180,
              ('I', 21): 220,
              ('I', 3): 250,
              ('II', 1): 300,
              ('II', 3): 250,
            ('II', 4): 240}   # em reais por hora
model.cproc = pyo.Param(model.Oleos,model.Centros, initialize=custo_processamento, default=0)

custo_materia_prima = {'I': 5, 'II': 6}   # em reais por litro
model.cmat = pyo.Param(model.Oleos, initialize=custo_materia_prima)

preco_venda = {'I': 20, 'II': 18}   # em reais por litro
model.preco = pyo.Param(model.Oleos, initialize=preco_venda)

venda_diaria = {'I': 1700, 'II': 1500}   # em litros
model.vdiaria = pyo.Param(model.Oleos, initialize=venda_diaria)

horas_funcionamento = {1: 16, 2: 12, 3: 12, 4: 16, 21: 12}  # em horas por dia
model.horas = pyo.Param(model.Centros, initialize=horas_funcionamento)

model.cap_transporte = pyo.Param(initialize=2500)  # em litros por dia
```

```python
# 3) Variáveis de decisão
model.x = pyo.Var(model.Oleos, model.Centros, domain=pyo.NonNegativeReals)  # litros de tipo de óleo processados em cada centro
```

```python
# 4) Função objetivo
def lucro(model):
    return (model.preco['I'] * (0.8*model.x['I', 21] + 0.75*model.x['I', 3]) + model.preco['II'] * 0.8*model.x['II', 4]) - \
        (model.cmat['I'] * model.x['I',1] + model.cmat['II'] * model.x['II',1]) - \
        sum((model.cproc[i,j] * model.x[i,j])/model.cap[i,j] for i in model.Oleos for j in model.Centros)

model.obj = pyo.Objective(rule=lucro, sense=pyo.maximize)
```

```python
# 5) Restrições
def capacidade_processamento(model, i, j):
    return model.x[i,j] <= model.cap[i, j]*model.horas[j]
model.capacidade_processamento = pyo.Constraint(model.Oleos, model.Centros, rule=capacidade_processamento)

def fluxos_recuperacao1(model):
    return model.x['I',2] == model.rec['I',1] * model.x['I',1]
model.fluxo_recuperacao1 = pyo.Constraint(rule=fluxos_recuperacao1)

def fluxos_recuperacao2(model):
    return model.x['I',4] == model.rec['I',2] * model.x['I',2]
model.fluxo_recuperacao2 = pyo.Constraint(rule=fluxos_recuperacao2)

def fluxos_recuperacao3(model):
    return model.x['I',21] + model.x['I',3] == model.rec['I',4] * model.x['I',4]
model.fluxo_recuperacao3 = pyo.Constraint(rule=fluxos_recuperacao3)

def fluxos_recuperacao4(model):
    return model.x['II',3] == model.rec['II',1] * model.x['II',1]
model.fluxo_recuperacao4 = pyo.Constraint(rule=fluxos_recuperacao4)

def fluxos_recuperacao5(model):
    return model.x['II',4] == model.rec['II',3] * model.x['II',3]
model.fluxo_recuperacao5 = pyo.Constraint(rule=fluxos_recuperacao5)

def fluxos_recuperacao6(model):
    return model.x['II',2] == 0
model.fluxo_recuperacao6 = pyo.Constraint(rule=fluxos_recuperacao6)

def fluxos_recuperacao7(model):
    return model.x['II',21] == 0
model.fluxo_recuperacao7 = pyo.Constraint(rule=fluxos_recuperacao7)

def venda_diaria_maxima_oleoI(model):
    return model.rec['I',21]*model.x['I',21] + model.rec['I',3]*model.x['I',3] <= model.vdiaria['I']
model.venda_diaria_oleoI = pyo.Constraint(rule=venda_diaria_maxima_oleoI)

def venda_diaria_maxima_oleoII(model):
    return model.rec['II',4]*model.x['II',4] <= model.vdiaria['II']
model.venda_diaria_oleoII = pyo.Constraint(rule=venda_diaria_maxima_oleoII)

def capacidade_transporte(model):
    return model.rec['I',21]*model.x['I',21] + model.rec['I',3]*model.x['I',3] + model.rec['II',4]*model.x['II',4] <= model.cap_transporte
model.capacidade_transporte = pyo.Constraint(rule=capacidade_transporte)

# Obs: não negatividade já está garantida pela definição das variáveis de decisão
```

```python
# 6) Solver
solver = pyo.SolverFactory('glpk')
results = solver.solve(model, tee=True)
```

```python
# 7) Resultados
print("Status da solução:", results.solver.status)
print("Condição de otimalidade:", results.solver.termination_condition)
```

```python
print("Lucro máximo: R$", pyo.value(model.obj))
```

```python
print("Quantidade de óleo de cada tipo processado em cada centro:")
for i in model.Oleos:
    for j in model.Centros:
        print(f"Óleo {i} no centro {j}: {pyo.value(model.x[i,j])} litros")
```

```python
print("Quantidade total de óleo produzido de cada tipo:")
print(f"Óleo I: {pyo.value(model.x['I',21])*pyo.value(model.rec['I', 21]) + pyo.value(model.x['I', 3])*pyo.value(model.rec['I', 3])} litros")
print(f"Óleo II: {pyo.value(model.x['II',4])*pyo.value(model.rec['II', 4])} litros")
```

## Modelagem alternativa

Agora vamos analisar uma abordagem alternativa de modelagem, conforme sugestão dos autores do livro de onde o problema foi extraído.

Inicialmente vamos definir a variável de decisão $x_i$ como o número de litros de óleo do tipo $i$ processados diariamente. Então desdobramos $x_I$ conforme o fluxo de processamento (N: natural; A: alternativo) da seguinte forma:
$$ x_I = x^N_I + x^A_I $$

Agora vamos modelar as perdas de matéria-prima em cada fluxo:
- Fluxo I Normal
$$ x_I \xrightarrow{\text{Centro 1}} 0.9 \cdot x_I \xrightarrow{\text{Centro 2}} 0.9 \cdot 0.95 \cdot x_I \xrightarrow{\text{Centro 4}} 0.9 \cdot 0.95 \cdot 0.85 \cdot x_I \xrightarrow{\text{Centro 2}} 0.9 \cdot 0.95 \cdot 0.85 \cdot 0.8 \cdot x^N_I$$
- Fluxo I Alternativo
$$ x_I \xrightarrow{\text{Centro 1}} 0.9 \cdot x_I \xrightarrow{\text{Centro 2}} 0.9 \cdot 0.95 \cdot x_I \xrightarrow{\text{Centro 4}} 0.9 \cdot 0.95 \cdot 0.85 \cdot x_I \xrightarrow{\text{Centro 3}} 0.9 \cdot 0.95 \cdot 0.85 \cdot 0.75 \cdot x^A_I$$
- Fluxo II
$$ x_{II} \xrightarrow{\text{Centro 1}} 0.9 \cdot x_{II} \xrightarrow{\text{Centro 3}} 0.9 \cdot 0.85 \cdot x_{II} \xrightarrow{\text{Centro 4}} 0.9 \cdot 0.85 \cdot 0.8 \cdot x_{II}$$

Restrições de capacidade de processamento:
- Centro 1:
$$ \frac{x^A_I}{300} + \frac{x^N_I}{300} + \frac{x_{II}}{500} \leq 16 $$
- Centro 2:
$$ \frac{0.9 \cdot x^N_I}{450} + \frac{0.9 \cdot x^A_I}{450} + \frac{0.9 \cdot 0.95 \cdot 0.85 \cdot x^N_I}{400} \leq 12 $$
- Centro 3:
$$ \frac{0.9 \cdot x_{II}}{480} + \frac{0.9 \cdot 0.95 \cdot 0.85 \cdot x^A_I}{350} \leq 12 $$
- Centro 4:
$$ \frac{0.9 \cdot 0.95 \cdot x^N_I}{250} + \frac{0.9 \cdot 0.95 \cdot x^A_I}{250} + \frac{0.9 \cdot 0.85 \cdot x_{II}}{400} \leq 16 $$

Restrição de transporte:
$$ 0.9 \cdot 0.95 \cdot 0.85 \cdot 0.8 \cdot x^N_I + 0.9 \cdot 0.95 \cdot 0.85 \cdot 0.75 \cdot x^A_I + 0.9 \cdot 0.85 \cdot 0.8 \cdot x_{II} \leq 2500 $$

Restrição de venda:
- Óleo tipo I:
$$ 0.9 \cdot 0.95 \cdot 0.85 \cdot 0.8 \cdot x^N_I + 0.9 \cdot 0.95 \cdot 0.85 \cdot 0.75 \cdot x^A_I \leq 1700 $$
- Óleo tipo II:
$$ 0.9 \cdot 0.85 \cdot 0.8 \cdot x_{II} \leq 1500 $$

Função objetivo:
$$ \max Z = \text{Receita} - \text{Custo de matéria-prima} - \text{Custos operacionais} $$
- Receita:
$$ 20 \cdot (0.9 \cdot 0.95 \cdot 0.85 \cdot 0.8 \cdot x^N_I) + 20 \cdot (0.9 \cdot 0.95 \cdot 0.85 \cdot 0.75 \cdot x^A_I) + 18 \cdot (0.9 \cdot 0.85 \cdot 0.8 \cdot x_{II}) $$
- Custo de matéria-prima:
$$ 5 \cdot x^N_I + 5 \cdot x^A_I + 6 \cdot x_{II} $$
- Custos operacionais:
    - Centro 1:
    $$ \frac{150 \cdot x^N_I}{300} + \frac{150 \cdot x^A_I}{300} + \frac{300 \cdot x_{II}}{500} $$
    - Centro 2:
    $$ \frac{200 \cdot 0.9 \cdot x^N_I}{450} + \frac{200 \cdot 0.9 \cdot x^A_I}{450} + \frac{220 \cdot 0.9 \cdot 0.95 \cdot 0.85 \cdot x^N_I}{400} $$
    - Centro 3:
    $$ \frac{250 \cdot 0.9 \cdot x_{II}}{480} + \frac{250 \cdot 0.9 \cdot 0.95 \cdot 0.85 \cdot x^A_I}{350} $$
    - Centro 4:
    $$ \frac{180 \cdot 0.9 \cdot 0.95 \cdot x^N_I}{250} + \frac{180 \cdot 0.9 \cdot 0.95 \cdot x^A_I}{250} + \frac{240 \cdot 0.9 \cdot 0.85 \cdot x_{II}}{400}

```python
model2 = pyo.ConcreteModel()
```

```python
# 1) Conjuntos
model2.Oleos = pyo.Set(initialize=["IN", "IA", "II"])
model2.Centros = pyo.Set(initialize=[1,2,3,4])
```

```python
# 2) Variáveis de decisão
model2.x = pyo.Var(model2.Oleos, model2.Centros, domain=pyo.NonNegativeReals)  # litros de tipo de óleo processados em cada centro
```

```python
# 3) Restrições
def capacidade_centro1(model2):
    return model2.x['IA',1]/300 + model2.x['IN', 1]/300 + model2.x["II", 1]/500 <= 16
model2.capacidade_centro1 = pyo.Constraint(rule=capacidade_centro1)

def capacidade_centro2(model2):
    return 0.9*model2.x['IA',1]/450 + 0.9*model2.x['IN', 1]/450 + 0.9*0.95*0.85*model2.x["IN", 1]/400 <= 12
model2.capacidade_centro2 = pyo.Constraint(rule=capacidade_centro2)

def capacidade_centro3(model2):
    return 0.9*model2.x['II',1]/480 + 0.9*0.95*0.85*model2.x["IA", 1]/350 <= 12
model2.capacidade_centro3 = pyo.Constraint(rule=capacidade_centro3)

def capacidade_centro4(model2):
    return 0.9*0.95*model2.x['IN',1]/250 + 0.9*0.95*model2.x["IA", 1]/250 + 0.9*0.85*model2.x["II", 1]/400 <= 16
model2.capacidade_centro4 = pyo.Constraint(rule=capacidade_centro4)

def transporte(model2):
    return 0.9*0.95*0.85*0.8*model2.x['IN',1] + 0.9*0.95*0.85*0.75*model2.x['IA',1] + 0.9*0.85*0.8*model2.x['II',1] <= 2500
model2.transporte = pyo.Constraint(rule=transporte)

def venda_oleoI(model2):
    return 0.9*0.95*0.85*0.8*model2.x['IN',1] + 0.9*0.95*0.85*0.75*model2.x['IA',1] <= 1700
model2.venda_oleoI = pyo.Constraint(rule=venda_oleoI)

def venda_oleoII(model2):
    return 0.9*0.85*0.8*model2.x['II',1] <= 1500
model2.venda_oleoII = pyo.Constraint(rule=venda_oleoII)
```

```python
# 4) Função objetivo
def lucro(model2):
    receita = 20*(0.9*0.95*0.85*0.8*model2.x['IN',1]) + 20*(0.9*0.95*0.85*0.75*model2.x['IA',1]) + 18*(0.9*0.85*0.8*model2.x['II',1])
    custo_materia_prima = 5*model2.x['IN',1] + 5*model2.x['IA',1] + 6*model2.x['II',1]
    custo_centro1 = (150*model2.x['IN',1])/300 + (150*model2.x['IA',1])/300 + (300*model2.x['II',1])/500
    custo_centro2 = (200*0.9*model2.x['IN',1])/450 + (200*0.9*model2.x['IA',1])/450 + (0.9*0.95*0.85*220*model2.x['IN',1])/400
    custo_centro3 = (250*0.9*0.95*0.85*model2.x['IA',1])/350 + (0.9*250*model2.x['II',1])/480
    custo_centro4 = (0.9*0.95*180*model2.x['IN',1])/250 + (0.9*0.95*180*model2.x['IA',1])/250 + (0.9*0.85*240*model2.x['II',1])/400
    return receita - custo_materia_prima - custo_centro1 - custo_centro2 - custo_centro3 - custo_centro4    

model2.obj = pyo.Objective(rule=lucro, sense=pyo.maximize)
```

```python
# 6) Solver
solver2 = pyo.SolverFactory('glpk')
results2 = solver2.solve(model2, tee=True)
```

```python
# 7) Resultados
print("Status da solução:", results2.solver.status)
print("Condição de otimalidade:", results2.solver.termination_condition)
```
